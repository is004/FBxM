```python
def render_word_cloud_topic(model, rows, cols, max_words, cmap, savepath):
  word_cloud = WordCloud(background_color='white', max_font_size=60, max_words=max_words, prefer_horizontal=1.0, colormap = cmap, random_state = 1)
  fig, axes = plt.subplots(rows, cols, figsize=(30,8))

  for i, ax in enumerate(axes.flatten()):
      fig.add_subplot(ax)
      topic_words = dict(model.show_topic(i))
      word_cloud.generate_from_frequencies(topic_words)
      plt.gca().imshow(word_cloud, interpolation='bilinear')
      plt.gca().set_title('Grup Topik {id}'.format(id=i+1))
      plt.gca().axis('off')

  plt.axis('off')
  plt.savefig(savepath, format = 'png', bbox_inches='tight')
  plt.show()
```

Jaccard score (Jaccard index)

```python
def jaccard(list1, list2):
    # intersection of two sets
    intersection = len(set(list1).intersection(list2))
    # Unions of two sets
    union = len(set(list1).union(list2))

    return round(intersection/union*100, 2)
```

#Pemrosesan mutual keywords & perhitungan coverage

Data preparation

```python
# Buat list of set df_keywords
# input = df_*_keywords
# output = list_set_*_keywords

def make_keyword_list_set(list_in):
  list_out = []
  for i in range (0, len(list_in)):
    temp_list = set(''.join(list_in[i]).replace(',', '').split())
    list_out.append(temp_list)
  return list_out

# Buat list of set df_contents
# input = df_*_contents
# output = list_set_*_contents

def make_content_list_set(list_in):
  list_out = []
  for i in range (0, len(list_in)):
    temp_list = set(''.join(list_in[i]).replace(',', '').replace("'", "").replace('[', '').replace(']', '').split())
    list_out.append(temp_list)
  return list_out
```

Pemrosesan Mutual Keywords

```python
# Buat mutual keywords list
# input1 = list_set_*_keywords
# input2 = list_set_*_contents
# output = mutual_set_list

def mutual_keywords_list(list_keywords, list_contents):
  mutual_set_list = []
  for i in range (0, len(list_keywords)):
    if len(list_keywords[i] & list_contents[i]) == 0:
      mutual_set_list.append('')
    else:
      mutual_set_list.append(list_keywords[i] & list_contents[i])
  return mutual_set_list
```

```python
# Hitung & buat list mutual keywords count
# input = mutual_set_list
# output = count of mutual_set_list list

def mutual_keywords_list_count(mutual_set_list):
  mutual_count = []
  for i in range (0, len(mutual_set_list)):
    mutual_count.append(len(mutual_set_list[i]))
  return mutual_count
```

```python
# Hitung Jaccard score individual
# input1 = list_set_*_keywords
# input2 = list_set_*_contents
# output = Jaccard score list

def jaccard(list_keywords_each, list_contents_each):
    # intersection of two sets
    intersection = len(list_keywords_each.intersection(list_contents_each))
    # Unions of two sets
    union = len(list_keywords_each.union(list_contents_each))
    return round(intersection/union*100, 2)

# Buat list Jaccard score seluruh individuals
def jaccard_list(list_keywords, list_contents):
  j_list = []
  for i in range (0, len(list_keywords)):
    j_list.append(jaccard(list_keywords[i], list_contents[i]))
  return j_list
```

Hitung coverage & false rate
<br><br>
Coverage = true_count = dokumen content yang berhasil dikelompokkan ke dalam salah satu topik (intersection != 0) = Jaccard score > 0
<br>
False rate = false_count = dokumen content yang gagal dikelompokkan ke dalam salah satu topik (intersection = 0) = Jaccard score = 0

```python
def intersection_coverage(mutual_keyword_count):
  true_count = 0
  false_count = 0
  for i in range (0, len(mutual_keyword_count)):
    if mutual_keyword_count[i] != 0:
      true_count += 1
    else:
      false_count += 1
    coverage = round(100*true_count/len(mutual_keyword_count), 2)
    false_rate = round(100*false_count/len(mutual_keyword_count), 2)
  return coverage, false_rate
```

Distribusi dokumen pada masing-masing topik

```python
#Hitung jumlah dokumen pada masing-masing topik
def count_topic_doc (topic_data_list):
  count_list = []
  label_list = []
  for i in range (0, max(topic_data_list)+1):
    count_list.append(topic_data_list.count(i))
    label_list.append('Topik {}'.format(i+1))
  return count_list, label_list
```
